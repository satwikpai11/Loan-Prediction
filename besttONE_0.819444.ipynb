```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv("train2.csv", delimiter=',')
df_train.head()
```

```python
a = list(df_train.index)

for i in a:
    if (df_train['Loan_Status'].iloc[i]=='Y'):
        df_train['Loan_Status'].iloc[i]='1'
    else:
        df_train['Loan_Status'].iloc[i]='0'
```

```python
df_train = df_train.astype({'Loan_Status': 'int'})
```

```python
df_train=df_train.drop(['Loan_ID'], axis=1)
```

```python
df_train = pd.get_dummies(df_train)
len(list(df_train.columns))
```

```python
df_train.columns
```

```python
x = df_train.drop(['Loan_Status'],axis=1)
y = df_train['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
import xgboost as xgb

train = xgb.DMatrix(X_train, label=y_train)
test = xgb.DMatrix(X_test)
```

```python
param = {
    'max_depth': 7,
    'eta': 0.2,

    'objective': 'multi:softmax',
    'num_class': 3} 
epochs = 7
```

```python
model = xgb.train(param, train, epochs)
```

```python
predictions = model.predict(test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```
